```python
import pandas as pd
import numpy as np

# Cantidad de datos
n = 200

# Creamos las variables cualitativas
Tipo_Energia = np.random.choice(['Energía Renovable', 'Energía No Renovable'], n, replace=True)
Politica_Ambiental = np.random.choice(['Alta Regulación', 'Moderada Regulación', 'Baja Regulación'], n, replace=True)
Conciencia_Social = np.random.choice(['Alta Conciencia', 'Media Conciencia', 'Baja Conciencia'], n, replace=True)
# Inicializamos las variables cuantitativas
Emisiones_CO2 = np.zeros(n)
Deforestacion = np.zeros(n)
Uso_Energia = np.zeros(n)
Inversion_Renovable = np.zeros(n)

# Generamos las variables cuantitativas condicionadas a las cualitativas
for i in range(n):
    # Condicionamos las emisiones de CO2
    if Tipo_Energia[i] == "Energía No Renovable":
        Emisiones_CO2[i] = np.round(np.random.uniform(30000, 50000))
    else:
        Emisiones_CO2[i] = np.round(np.random.uniform(1000, 15000))

    # Condicionamos la deforestación
    if Politica_Ambiental[i] == "Baja Regulación":
        Deforestacion[i] = np.round(np.random.uniform(5000, 10000))
    elif Politica_Ambiental[i] == "Moderada Regulación":
        Deforestacion[i] = np.round(np.random.uniform(2000, 5000))
    else:
        Deforestacion[i] = np.round(np.random.uniform(100, 2000))

    # Condicionamos el uso de energía
    if Tipo_Energia[i] == "Energía No Renovable":
        Uso_Energia[i] = np.round(np.random.uniform(50000, 100000))
    else:
        Uso_Energia[i] = np.round(np.random.uniform(1000, 50000))

    # Condicionamos la inversión en energía renovable
    if Conciencia_Social[i] == "Alta Conciencia":
        Inversion_Renovable[i] = np.round(np.random.uniform(3000, 5000))
    elif Conciencia_Social[i] == "Media Conciencia":
        Inversion_Renovable[i] = np.round(np.random.uniform(1000, 3000))
    else:
        Inversion_Renovable[i] = np.round(np.random.uniform(10, 1000))

# Creamos el dataframe con las variables
base_cambio_climatico = pd.DataFrame({
    'Tipo_Energia': Tipo_Energia,
    'Politica_Ambiental': Politica_Ambiental,
    'Conciencia_Social': Conciencia_Social,
    'Emisiones_CO2': Emisiones_CO2,
    'Deforestacion': Deforestacion,
    'Uso_Energia': Uso_Energia,
    'Inversion_Renovable': Inversion_Renovable
})
# Guardamos el dataframe como un archivo CSV
base_cambio_climatico.to_csv('C:/Users/57302/OneDrive - UCO/Cambio climatico/Base_Cambio_Climatico.csv', index=False)
```

```python
columnas=list(df.select_dtypes(include=['float']).columns)
for x in columnas:
    skewness=df[x].skew()
    Q1=df[x].quantile(0.25)
    Q3=df[x].quantile(0.75)
    IQR=Q3-Q1
    std_val=df[x].std()
    mean_val=df[x].mean()
    cv=(std_val/mean_val)*100
    print(f"para la columna {x} el valor de rango intercuartilico es {IQR}")
    print(f"para la columna {x} el valor de coeficiente de variacion es {cv}")
    print(f"para la columna {x} el valor de skewness es {skewness}")
```

```python
# Crear una lista de datos numéricos
datos = df[columnas_numericas[0]]

# Definir el número de intervalos usando la regla de Sturges
n = len(datos)
num_intervalos = int(1 + 3.3 * np.log10(n))

# Crear los intervalos
rango_min = min(datos)
rango_max = max(datos)
intervalos = np.linspace(rango_min, rango_max, num_intervalos + 1)  # Creamos los límites de los intervalos

# Crear la tabla de frecuencias
tabla_frecuencia = pd.cut(datos, bins=intervalos, right=False)  # 'right=False' excluye el límite derecho del intervalo
frecuencias = pd.value_counts(tabla_frecuencia, sort=False)  # Obtener frecuencias absolutas

# Crear un DataFrame con los intervalos y frecuencias
df_frecuencias = pd.DataFrame({
    'Intervalo': frecuencias.index,
    'Frecuencia': frecuencias.values
})

# Calcular la frecuencia relativa y acumulada
df_frecuencias['Frecuencia Relativa'] = df_frecuencias['Frecuencia'] / n
df_frecuencias['Frecuencia Acumulada'] = df_frecuencias['Frecuencia'].cumsum()
df_frecuencias['Frecuencia Relativa Acumulada']=df_frecuencias["Frecuencia Relativa"].cumsum()
print(columnas_numericas[0])
print(df_frecuencias)
```
